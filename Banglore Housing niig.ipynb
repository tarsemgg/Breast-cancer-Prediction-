```python
def floats(x):
    try:
        float(x)
    except:
        return False
    return True
# this peice of code is exactly what it looks like here i defined a function to make all values fload of a perticular coloumn
```

```python
df2[~df2['total_sqft'].apply(floats)].head(30)
#this returns the rows which does not fullfill the requerements of above funtion and that is due to (~) negate functun
```

```python
def the_mean(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return (float(tokens[0])+ float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None 
```

```python
df3=df2.copy()
df3['total_sqft']= df2['total_sqft'].apply(the_mean)
```

```python
def to_sqft1(x):
    if x.endswith('Perch'):
        perch_val=float(x[-6])
        sqft_val= perch_val*272.25
        return sqft_val
    try:
        return float(x)
    except:
        return None
```

```python
df4=df3.copy()
df4['total_sqft']= df2['total_sqft'].apply(to_sqft1)
```

```python
df4.iloc[648]
```

```python
def to_sqft2(x):
    if x.endswith('Sq.Yards'):
        yards= float(x[-8])
        sqft= yards*9.0
        return sqft
    try: 
        return float(x)
    except:
        return None
```

```python
df5=df4.copy()
df5['total_sqft']= df2['total_sqft'].apply(to_sqft2)
```

```python
def to_sqft3(x):
    if x.endswith('Acres'):
        acres= float(x[-6])
        sqft= acres*43560.0
        return sqft
    try:
        return float(x)
    except:
        return None
```

```python
def price_pred1(location, sqft, bath, bhk):
    index_loc = np.where(x.columns == location)[0][0]
    feature_vector = np.zeros(len(x.columns))
    feature_vector[0] = sqft
    feature_vector[1] = bath
    feature_vector[2] = bhk
    if index_loc >= 0:
        feature_vector[index_loc] = 1
    return model.predict([feature_vector])[0]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV

models = {
    'Linear Regression': {
        'model': LinearRegression(),
        'params': {}
    },
    'Random Forest Regressor': {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [10, 50, 100],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5, 10]
        }
    },
    'Support Vector Regressor': {
        'model': SVR(),
        'params': {
            'kernel': ['linear', 'poly', 'rbf'],
            'C': [0.1, 1, 10]
        }
    }
}
best_model = None
best_params = None
best_score = float('inf')

for model_name, model_info in models.items():
    model = model_info['model']
    params = model_info['params']
    
    grid_search = GridSearchCV(model, params, scoring='neg_mean_squared_error', cv=5)
    grid_search.fit(xtrain, ytrain)
    
    if grid_search.best_score_ < best_score:
        best_model = model_name
        best_params = grid_search.best_params_
        best_score = grid_search.best_score_

print(f'Best Regression Model: {best_model}')
print(f'Best Parameters: {best_params}')
```
